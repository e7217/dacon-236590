### 데이터 분포 확인

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기 (파일 경로는 실제 환경에 맞게 수정해주세요)
train_df = pd.read_csv('./data/open/train.csv')

# target 변수의 분포 확인 (각 클래스별 개수 출력)
print(train_df['target'].value_counts())

# 분포 시각화
plt.figure(figsize=(10, 6))
sns.countplot(x='target', data=train_df)
plt.title('Target Distribution')
plt.show()
```

데이터 분포가 동일함을 확인하였다

```python
# 1. 데이터 기본 정보 확인 (결측치, 데이터 타입 등)
print("--- Data Info ---")
train_df.info()

print("\n" + "="*50 + "\n")

# 2. 기초 통계량 확인
print("--- Descriptive Statistics ---")
display(train_df.describe())
display(train_df.describe().T)
```

### 데이터 정규화

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# X는 이전에 target을 제외한 피처 데이터프레임입니다.
scaler = MinMaxScaler()

# 스케일러를 데이터에 학습(fit)시키고 변환(transform)합니다.
X = train_df.drop(columns=['ID', 'target'])
X_scaled = scaler.fit_transform(X)

# 결과를 다시 데이터프레임으로 만들어줍니다.
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# 변환된 결과의 통계 정보를 확인해봅시다.
X_scaled.describe().T
```

### 기본 성능 확인
로직스틱 회귀를 이용해서 기본 성능 확인

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

# 1. 데이터를 훈련용과 시험용으로 분리합니다.
# (모델의 진짜 실력을 객관적으로 평가하기 위함)
y = train_df['target']
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 2. 모델을 만들고 훈련용 데이터로 학습시킵니다.
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# 3. 학습된 모델로 시험용 데이터의 정답을 예측합니다.
predictions = model.predict(X_test)

# 4. Macro F1-Score로 예측 성능을 채점합니다.
score = f1_score(y_test, predictions, average='macro')
print(f"기준 모델의 Macro F1-Score: {score}")
```

### EDA 진행

```python
# train_df는 target과 X 피처들이 모두 포함된 원본 데이터프레임입니다.
# X_01 피처가 target 별로 어떻게 분포하는지 확인
plt.figure(figsize=(12, 8))
sns.boxplot(x='target', y='X_01', data=train_df)
plt.title('Box Plot of X_01 by Target')
plt.show()
```

```python
import lightgbm as lgb

# 1. LightGBM 모델 생성 및 학습
lgbm = lgb.LGBMClassifier(device='gpu', random_state=42)
lgbm.fit(X_train, y_train)

# 2. 성능 평가 (우리 기준 점수인 0.512와 비교해보세요)
predictions = lgbm.predict(X_test)
score = f1_score(y_test, predictions, average='macro')
print(f"LightGBM 모델의 Macro F1-Score: {score}")

# 3. 피처 중요도 확인 및 시각화
feature_importances = pd.Series(lgbm.feature_importances_, index=X_train.columns)
feature_importances = feature_importances.sort_values(ascending=False)

plt.figure(figsize=(10, 12))
sns.barplot(x=feature_importances, y=feature_importances.index)
plt.title('Feature Importances')
plt.show()
```
